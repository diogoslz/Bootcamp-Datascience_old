### Desafio 01: Escolher uma palete de cores mais adequada do matplotlib.

```python
import random

random.seed(1000)

cores = []

for i in range(27):
    r = lambda: random.randint(0,255)
    
    '#%02X%02X%02X' % (r(),r(),r())
    
    cores.append('#%02X%02X%02X' % (r(),r(),r()))
```

```python
customPalette = sns.set_palette(sns.color_palette(cores))

plt.figure(figsize=(12,8))
ax = sns.lineplot(data=usaveis.drop(['Total','Região'], axis=1).T, dashes=False, palette= customPalette)

ax.legend(bbox_to_anchor = (1, 1))
plt.show()
```
