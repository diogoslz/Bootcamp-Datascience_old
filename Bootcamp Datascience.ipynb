```python
import pandas as pd
```

```python
dados = pd.read_csv('A151346189_28_143_208.csv', encoding='ISO-8859-1', sep=';', skiprows=3,
                    skipfooter=12, thousands='.', decimal=',')
```

```python
dados.head()
```

```python
dados.mean()
```

```python
pd.options.display.float_format = '{:,.2f}'.format
```

```python
dados['2008/Jan']
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title("Valor por unidade da federação")
plt.show()
```

```python
### Desafio 01: Escolher um título mais descritivo, que passe a mensagem adequada ao gráfico de barras.
```

```python
dados.sort_values('2008/Ago', ascending=False)
```

```python
ax = dados.sort_values('2008/Ago', ascending=False).plot(x='Unidade da Federação', y='2008/Ago', kind="bar", figsize=(9,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title("Valores Gasto da Saúde por unidade da federação (R$)")
plt.show()
```
