```python
import pandas as pd
```

```python
dados = pd.read_csv('A225850189_28_143_208.csv', encoding='ISO-8859-1', sep=';', skiprows=3,
                    skipfooter=12, thousands='.', decimal=',')
```

```python
dados.head()
```

```python
dados.mean()
```

```python
pd.options.display.float_format = '{:,.2f}'.format
```

```python
dados['2008/Jan']
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title("Valor por unidade da federação")
plt.show()
```

```python
### Desafio 01: Escolher um título mais descritivo, que passe a mensagem adequada ao gráfico de barras.
```

```python
ax = dados.sort_values('2008/Ago', ascending=False).plot(x='Unidade da Federação', y='2008/Ago', kind="bar", figsize=(9,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title("Valores Gasto da Saúde por unidade da federação (R$)")
plt.show()
```

```python
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0,"Unidade da Federação")
colunas_usaveis
```

```python
usaveis = dados[colunas_usaveis]
usaveis.head()
```

```python
usaveis = usaveis.set_index("Unidade da Federação")
usaveis.head()
```

```python
usaveis["2019/Ago"].head()
```

```python
usaveis.loc["12 Acre"]
```

```python
usaveis.T.head()
```

```python
ax = usaveis.T.plot(figsize=(10,6))
plt.show()
```

```python
usaveis = usaveis.drop("Total", axis=1)
```

```python
##Desafio 01: Reposicionar a legenda do gráfico em uma posição mais adequada
```

```python
ax = usaveis.T.plot(figsize=(10,7.5))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
ax.legend(bbox_to_anchor = (1, 1))
plt.show()
```

```python
##Desafio 02: Plotar o gráfico de linha com apenas 5 estados de sua preferência
```
