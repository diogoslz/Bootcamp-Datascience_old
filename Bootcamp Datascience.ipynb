### Desafio 05: Buscar os casos de dengue no Brasil (época de maior número de casos e regiões mais atingidas) e se os picos de alguns estados em fevereiro e verão de modo geral, pode ser reflexos dos casos de dengue

```python
mypath = '.\\Dengue\\'
```

```python
from os import listdir
from os.path import isfile, join
arquivos_dengue = [f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
arquivos_dengue
```

```python
f[:4]
```

```python
list_base_dengue = []
index = 0

for f in arquivos_dengue:
    
    
    if (index <= 5): 
    
        df = pd.read_csv(mypath + f, encoding='ISO-8859-1', sep=';', skiprows=3, skipfooter=17, thousands='.', decimal=',')
        
    elif (index == 8):
        
        df = pd.read_csv(mypath + f, encoding='ISO-8859-1', sep=';', skiprows=3, skipfooter=20, thousands='.', decimal=',')

    else:
        
        df = pd.read_csv(mypath + f, encoding='ISO-8859-1', sep=';', skiprows=3, skipfooter=19, thousands='.', decimal=',')
 
    
    estados = df['UF de notificação']
    
    df.drop('Total', axis = 1, inplace=True )
    
    df.drop('UF de notificação', axis = 1, inplace=True )
    
    df.columns = f[:4] + '/' + df.columns

    index += 1
    
    
    list_base_dengue.append(df) 
```

```python
df.columns
```

```python
dados_dengue = pd.concat (list_base_dengue, axis=1)
dados_dengue['Unidade da Federação'] = estados
dados_dengue.head()
```

```python
dengue_usaveis = dados_dengue.mean().index.tolist()
dengue_usaveis.insert(0,"Unidade da Federação")
dengue_usaveis
```

```python
dengue_usaveis = dados_dengue[dengue_usaveis]
dengue_usaveis.head()
```

```python
dengue_usaveis.set_index('Unidade da Federação', inplace=True)
```
